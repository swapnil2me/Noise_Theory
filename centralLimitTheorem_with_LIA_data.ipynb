```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataDir = 'testData'
for file in os.listdir(dataDir):
    if file.endswith('.csv'):
        print(file)
        df = pd.read_csv(os.path.join(dataDir,file))
        df.hist(column='A')
        plt.title('.'.join(file.split('.')[0:-1]))
```
