```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
np.random.randn(2,4)
```

```python
peakHeight = 1
peakPosition = 100
peakWidth = 40
smoothWidth = 60
noiseLevel = 0.01
frequency = 6
phaseShift = 0
```

```python
x = np.arange(1,200+0.1,0.1)
# White Noise
noise = noiseLevel * np.random.randn(*x.shape)
```

```python
plt.plot(noise,'.')
plt.title("White Noise")
```

```python
# random walk = cumulative sum of white noise
baseLine = noise * 10 + np.cumsum(noise)
```

```python
plt.plot(np.cumsum(noise),'.')
```

```python
plt.plot(baseLine,'.')
plt.title("Random Walk")
```

```python
y = peakHeight * norm.pdf(x, peakPosition, peakWidth)
```

```python
plt.plot(x,y)
```

```python
snr = peakHeight/np.std(baseLine)
oy = y + baseLine
```

```python
plt.plot(x,oy,'.')
```
